```python
import pandas as pd 

df_clean = pd.read_csv('../clean_dataset.csv')
df_clean
```

```python
# words and sequences that cab be used to filter our ads using scarcity
scarcity_words = [ 'limited time only', 'limited time', 'almost gone', 'only for' , 'limited edition', 'before it', 'last chance', 
                      'days left', 'missing out', 'now on sale', 'too late', 'almost run', 'weeks left', 'hours left','hurry', 'do not wait', 
                      'urge' , 'urging','almost done', 'desperate need', 'act now', 'no time', 'in a race against time', 'one of a kind', 
                      'waste time', 'ends soon', 'race against time', 'time running', 'urgence', 'immediately', 'limited', 'end soon' , 
                      'expire soon', 'expire', 'expires','deadline', 'seize', 'bargain', 'emergency', 'critical', 'hurry', 'last call', 
                      'urgently', 'urgent', 'less than','until', 'it is time','miss out', 'now is the time','now more than ever', 'do not miss',
                      'there is no time','time is running','running out of time','we do not have time',"before", "take action today",  "it is time",
                       "you will only have seven days", "it is high time for a change", "chip in today","it is time for action","right now"]

def contains_word(s, w):
    if w in s.lower() :
      return True
    else: 
      return False
```

```python
scarcity_true, scarcity_false = set(), set()

for sentence in df_clean['ads']:                                   
  for word in  scarcity_words:                                        
      if contains_word(sentence, word):
        scarcity_true.add(sentence)

for sentence in df_clean['ads']:                                   
    if sentence not in scarcity_true:
        scarcity_false.add(sentence) 

# show the filtering results 
print(len(scarcity_true))
print(len(scarcity_false))
```

```python
df_scarcity_true = pd.DataFrame(list(scarcity_true), columns= ['ads'])
df_scarcity_true['scarcity'] = 1
df_scarcity_true
```

```python
df_scarcity_false = pd.DataFrame(list(scarcity_false), columns= ['ads']).sample( n = 15000).reset_index(drop=True)
df_scarcity_false['scarcity'] = 0
df_scarcity_false
```

```python
df_scarcity = pd.concat([df_scarcity_true, df_scarcity_false], ignore_index=True)
df_scarcity
```

```python
df_scarcity.to_csv('../../code/scarcity/data/scarcity_dataset_corrected.csv', index=False)
```
