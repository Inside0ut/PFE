```python
import pandas as pd

df_clean = pd.read_csv('../clean_dataset.csv')
df_clean.shape
```

```python
consensus_words = [ 'society', 'community', 'popular', 'many of', 'everyone thinks', 'everyone knows', 
                    'everyone believes', 'thousands of', 'millions of', 'more than', 'endorsed', 'endorsement', 'of people', ] 
def contains_word(s, w):
  if w in s.lower() :
    return True
  else: 
    return False
```

```python
consensus_true = set()

for sentence in df_clean['ads']:                                   
  for word in  consensus_words:                                        
      if contains_word(sentence, word):
        consensus_true.add(sentence)
        
# show the filtering results 
len(consensus_true)
```

```python
df_consensus_true = pd.DataFrame(list(consensus_true), columns = ['ads'])
df_consensus_true['consensus'] = 1
df_consensus_true.shape
```

```python
all_consensus_words = [ 'unanimous', 'unison', 'common','citizen', 'republican', 'democrat', 
                        'most' , 'poeple', 'more', 'donors', 'participants', 'voter', 'someone',
                        'americans', 'all', 'everyone', 'canadians', 'youth', 'old', 'mother', 'families',
                        'us', 'we', 'all', 'million', 'thousand', 'hundred', 'large', 'population','we all', 'majority', 
                        'society', 'community', 'popular', 'many of', 'everyone thinks', 'everyone knows', 
                        'everyone believes', 'thousands of', 'millions of', 'more than', 'endorsed', 'endorsement'] 
```

```python
consensus_true, consensus_false = set(), set()

for sentence in df_clean.ads:                                   
  for word in  all_consensus_words:                                       
      if contains_word(sentence, word):
        consensus_true.add(sentence)

for sentence in df_clean.ads:                                
  if sentence not in consensus_true:
    consensus_false.add(sentence)
```

```python
df_consensus_false = pd.DataFrame(list(consensus_false), columns = ['ads']).sample(n = 10000).reset_index(drop = True)
df_consensus_false.drop_duplicates(subset = 'ads', keep = 'first', inplace = True)
df_consensus_false['consensus'] = 0
df_consensus_false
```

```python
df_consensus = pd.concat([df_consensus_true, df_consensus_false], ignore_index = True).sample(frac = 1).reset_index(drop = True)
df_consensus 
```

```python
df_consensus.to_csv('../../code/consensus/data/consensus_dataset_2.csv', index = False)
```
