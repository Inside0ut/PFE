```python
import pandas as pd

df_clean = pd.read_csv('../clean_dataset.csv')
df_clean
```

```python
moral_appeal_words = ['rights', 'equal', 'justice', 'protect', 'equality',  'social' , 'environment',  'popular', 'prohibition' ]

# verify if sentence contains the desired word
def contains_word(s, w):
    if w in s.lower() :
      return True
    else: 
      return False
```

```python
moral_appeal_true = set()

for sentence in df_clean['ads']:                                   
  for word in  moral_appeal_words:                                        
      if contains_word(sentence, word):
        moral_appeal_true.add(sentence)
        
# show the filtering results 
len(moral_appeal_true)
```

```python
df_moral_appeal_true = pd.DataFrame(list(moral_appeal_true), columns= ['ads']).sample(n = 10000).reset_index(drop=True)
df_moral_appeal_true['moral_appeal'] = 1
df_moral_appeal_true
```

```python
moral_appeal_true, moral_appeal_false = set(), set()

for sentence in df_clean.ads:                                   
  for word in  large_filter:                                     
      if contains_word(sentence, word):
        moral_appeal_true.add(sentence)

for sentence in df_clean.ads:                                
  if sentence not in moral_appeal_true:
    moral_appeal_false.add(sentence)
    
len(moral_appeal_false)
```

```python
df_moral_appeal_false = pd.DataFrame(list(moral_appeal_false), columns= ['ads'])
df_moral_appeal_false['moral_appeal'] = 0
df_moral_appeal_false
```

```python
df_moral_appeal = pd.concat([df_moral_appeal_true, df_moral_appeal_false], ignore_index=True)
df_moral_appeal.drop_duplicates(subset = 'ads', keep = 'first', inplace= True)
df_moral_appeal
```
