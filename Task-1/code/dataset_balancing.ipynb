```python
import pandas as pd 
# import ads dataset 
# the dataset contains commercial ads tagged as 0 and informational ads tagged as 1 
ad=pd.read_csv('./data/ads_dataset.csv')
ad.head()
```

```python
# import a politcal ads dataset
# link : https://www.propublica.org/datastore/dataset/political-advertisements-from-facebook
poly=pd.read_csv('./data/political_ads.csv')
poly.rename(columns = {'Unnamed: 0':'label'}, inplace = True)
poly.rename(columns = {'0':'text'}, inplace = True)
poly['label']=1
poly.head()
```

```python
# we select 6000 political ads tagged as infromational (label : 1) to balance our data
poly = poly.sample(frac = 1)
selected=poly[0:6000]
selected.shape
```

```python
vertical_concat = pd.concat([ad,selected], axis=0)
vertical_concat = vertical_concat.sample(frac = 1)
vertical_concat.head()
```

```python
vertical_concat.shape
```

```python
index_names = vertical_concat[vertical_concat['label']== 2 ].index
  
# drop row indexes from dataFrame
vertical_concat.drop(index_names, inplace = True)
  
vertical_concat.shape
```

```python
vertical_concat = vertical_concat.dropna(subset=['text'])
vertical_concat.shape
```

```python
vertical_concat.label.unique()
```
